```python
import pandas as pd
```

```python
#data = pd.read_csv("datasets/ds_full_pt.csv")
data = pd.read_csv("datasets2/ds_full.csv")
data.head()
```

```python
X = data.drop('rotulo', axis=1)
Y = data['rotulo']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=109) 
```

```python
X_train
```

```python
df = pd.concat([X_train, y_train], axis=1, sort=False)  
df
```

```python
#df.to_csv('datasets/ds_full_treino_pt.csv', index=False)
df.to_csv('datasets2/ds_full_treino.csv', index=False)
```

```python
dfv = pd.concat([X_test, y_test], axis=1, sort=False)  
dfv
```

```python
#dfv.to_csv('datasets/ds_full_validacao_pt.csv', index=False)
dfv.to_csv('datasets2/ds_full_validacao.csv', index=False)
```
